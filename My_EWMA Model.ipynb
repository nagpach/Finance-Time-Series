```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
airline = pd.read_csv('airline_passengers.csv', index_col='Month')
```

```python
airline.head()
```

```python
airline.index
```

```python
airline.dropna(inplace=True)
```

```python
airline.index = pd.to_datetime(airline.index)
```

```python
airline.head()
```

```python
airline['6_month_SMA'] = airline['Thousands of Passengers'].rolling(window=6).mean()
```

```python
airline['12_month_SMA'] = airline['Thousands of Passengers'].rolling(window=12).mean()
```

```python
airline[['Thousands of Passengers', '6_month_SMA', '12_month_SMA']].plot(figsize=(12,6));
```

```python
airline['EWMA-12'] = airline['Thousands of Passengers'].ewm(span=12).mean()
```

```python
airline[['Thousands of Passengers', 'EWMA-12']].plot(figsize=(12,6));
```
